```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import math
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
#print(df)
v = df.values
f = v[:,:8] #Pregnancies,Glucose,BP,SkinThick,Insul,BMI,DPF,Age; n = 8
t = v[:,8] #Diabetic?
```

```python
X_features = ["Pregnancy","Glucose","BP","Skin Thickness","Insulin","BMI","DPF","Age"]
fig, ax = plt.subplots(1, 8, figsize= (12,3), sharey = True)
for i in range(len(ax)):
    ax[i].scatter(f[:,i],t)
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel("Diabetes")
plt.show()
```

```python
lr_model = LogisticRegression(max_iter=1000000000)
lr_model.fit(f,t)
```

```python
pred = lr_model.predict(f)

print("Prediction on training set:", pred)
```

```python
print("Accuracy on training set:", lr_model.score(f, t))
```
